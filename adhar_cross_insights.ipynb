```python
import pandas as pd

enrol_summary = pd.read_csv("enrol_summary.csv")
bio_summary   = pd.read_csv("bio_summary.csv")
demo_summary  = pd.read_csv("demo_summary.csv")
```

```python
cross_df = enrol_summary.merge(
    bio_summary, on="district", how="left"
).merge(
    demo_summary, on="district", how="left"
)

cross_df.fillna(0, inplace=True)
cross_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

cross_df[["enrol_norm", "bio_norm", "demo_norm"]] = scaler.fit_transform(
    cross_df[["total_enrolment", "total_biometric", "total_population"]]
)

cross_df["governance_risk_score"] = (
    cross_df["enrol_norm"] * 0.4 +
    cross_df["bio_norm"] * 0.4 +
    (1 - cross_df["demo_norm"]) * 0.2
)
```

```python
cross_df.sort_values(
    "governance_risk_score", ascending=False
).head(10)
```

```python
cross_df.to_csv("district_governance_risk_index.csv", index=False)
```
